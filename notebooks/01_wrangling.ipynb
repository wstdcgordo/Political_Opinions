```python
# Data Information
# Let's see Features :

# Cinsiyet : Sex Feature

# Yas : Age Feature

# Bolge : Areas inhabited in Turkey

# Egitim : Education Level

# (Soru = Question) (Questions include Turkey)

# Soru1/Question1: Do you think our Economic Status is good?

# Soru2/Question2: Do you think the education system needs to be reformed?

# Soru3/Question3: Do you support privatization?

# Soru4/Question4: Should the death penalty be used for certain crimes?

# Soru5/Question5: Do you think the media is neutral?

# Soru6/Question6: Do you support the alcohol ban after 22:00?

# Soru7/Question7: Do you want to live in a state where religion and state are separate?

# Soru8/Question8: Do you think abortion should be banned?

# Soru9/Question9: Do you think that the state of emergency restricts freedoms?

# Soru10/Question10: Do you want a new party to enter the parliament?

# Parti : Political View

# Data Review :

# Evet = Yes

# Hayır = No

# Erkek = Male

# Kadın = Female

# Yas = Age

# Bolge = City

# Egitim = Education

# İlkokul = primary school

# OrtaOkul = junior high school

# Lise = High School

# Lisans = University

# Lisans Üstü = MA
```

```python
# Concatenate three datasets
df1 = pd.read_csv(r'C:\Users\Ebo\Projetos Personais\Github Repositories\Political_Opinions\data\turkey-political-opinions\data.csv')
df2 = pd.read_csv(r'C:\Users\Ebo\Projetos Personais\Github Repositories\Political_Opinions\data\turkey-political-opinions\datav2.csv')
df3 = pd.read_csv(r'C:\Users\Ebo\Projetos Personais\Github Repositories\Political_Opinions\data\turkey-political-opinions\yonelimfinal.csv')

df = pd.concat([df1, df2, df3], axis=0).reset_index(drop=True)
```

```python
df.head(10)
```

```python
df.drop(columns=['Timestamp'], inplace=True)
```

```python
# The data is written in Turkish, so let's translate it to English.
# We would obtain unique values in some columns in need of translation, as well as the yes and no.

print(df['Cinsiyet'].unique())
print(df['Bolge'].unique())
print(df['Egitim'].unique())
print(df['parti'].unique())
```

```python
# Translate the columns to English
cinsiyet_map = {'Erkek': 'Male',
                'Kadın': 'Female'
                }

egitim_map = {'İlkokul': 'Primary School',
                'Ortaokul': 'Junior High School',
                'Lise': 'High School',
                'Lisans': 'Undergraduate',
                'Lisans Üstü': 'Graduate',
                'Ön Lisans': 'Associate'
                }

soru_map = {'Evet': 'Yes',
            'Hayır': 'No'
            }

parti_map = {'AKP': 'AKP',
            'CHP': 'CHP',
            'HDP': 'HDP',
            'IYI PARTI': 'IYI Parti',
            'MHP': 'MHP',
            'DIĞER': 'Other'
            }
            

df['Cinsiyet'] = df['Cinsiyet'].map(cinsiyet_map)
df['Egitim'] = df['Egitim'].map(egitim_map)
df[['soru1', 'soru2', 'soru3', 'soru4', 'soru5', 'soru6', 'soru7', 'soru8', 'soru9', 'soru10']] = df[['soru1', 'soru2', 'soru3', 'soru4', 'soru5', 'soru6', 'soru7', 'soru8', 'soru9', 'soru10']].applymap(lambda x: soru_map[x])
df['parti'] = df['parti'].map(parti_map)
```

```python
# Rename the cities into non-Turkish characters to avoid encoding issues and glitches
bolge_map = {'Marmara': 'Marmara',
             'Güneydoğu': 'Guneydogu',
             'Akdeniz': 'Akdeniz',
             'Doğu Anadolu': 'Dogu Anadolu',
             'İç Anadolu': 'Ic Anadolu',
             'Karadeniz': 'Karadeniz',
             'Ege': 'Ege'
             }

df['Bolge'] = df['Bolge'].map(bolge_map)
```

```python
df.head()
```

```python
# Rename the columns too, except the soru part
df.rename(columns={'Cinsiyet': 'Sex',
                   'Yas': 'Age',
                   'Bolge': 'City',
                   'Egitim': 'Education',
                   'parti': 'Political Affiliation'
}, inplace=True)
```

```python
# Saving into CSV with path
df.to_csv(r'C:\Users\Ebo\Github Repositories\Political_Opinions\data\01_wrangled.csv', index=False)
```

## Cleaning/Categorizing

```python
df_wrangled = pd.read_csv(r'C:\Users\Ebo\Github Repositories\Political_Opinions\data\01_wrangled.csv')
```

```python
df_both = df_wrangled.copy()
```

```python
df_both = df_both[df_both['Political Affiliation'] != 'Other']
df_other = df_wrangled[df_wrangled['Political Affiliation'] == 'Other']
```

```python
sex_mapping = {'Male': 0,
               'Female': 1}

age_mapping = {'0-18': 0,
               '18-30': 1,
               '30-50': 2,
               '50-60': 3,
               '60+': 4}

city_mapping = {'Marmara': 0,
                'Guneydogu': 1,
                'Akdeniz': 2,
                'Dogu Anadolu': 3,
                'Ic Anadolu': 4,
                'Karadeniz': 5,
                'Ege': 6
                }

education_mapping = {'Primary School': 0,
                        'Junior High School': 1,
                        'High School': 2,
                        'Undergraduate': 4,
                        'Graduate': 5,
                        'Associate': 3
                        }

spectrum_mapping = {'AKP': 1, # Right-leaning
                    'CHP': 0, # Left-leaning
                    'HDP': 0, # Left-leaning
                    'IYI Parti': 1, # Center-right
                    'MHP': 1 # Right-leaning
}

political_view_mapping = {'AKP': 1,  # Right-leaning
                          'CHP': 0,  # Left-Leaning
                          'HDP': 0,  # Left-leaning
                          'IYI Parti': 1,  # Right-leaning
                          'MHP': 1,  # Right-leaning
                          'Other': 2  # Other
}

soru_mapping = {'Yes': 1,
                'No': 0
}

df_both.loc[:, 'Political Affiliation'] = df_both['Political Affiliation'].map(political_view_mapping)
df_both.loc[:, 'Sex'] = df_both['Sex'].map(sex_mapping)
df_both.loc[:, 'Age'] = df_both['Age'].map(age_mapping)
df_both.loc[:, 'City'] = df_both['City'].map(city_mapping)
df_both.loc[:, 'Education'] = df_both['Education'].map(education_mapping)

df_other.loc[:, 'Political Affiliation'] = df_other['Political Affiliation'].map(political_view_mapping)
df_other.loc[:, 'Sex'] = df_other['Sex'].map(sex_mapping)
df_other.loc[:, 'Age'] = df_other['Age'].map(age_mapping)
df_other.loc[:, 'City'] = df_other['City'].map(city_mapping)
df_other.loc[:, 'Education'] = df_other['Education'].map(education_mapping)

for column in ['soru1', 'soru2', 'soru3', 'soru4', 'soru5', 'soru6', 'soru7', 'soru8', 'soru9', 'soru10']:
    df_both.loc[:, column] = df_both[column].map(soru_mapping)
    df_other.loc[:, column] = df_other[column].map(soru_mapping)
```

```python
# Rename the questions into their contexts
df_both.rename(columns={'soru1': 'Economic Status',
                   'soru2': 'Education Reform',
                   'soru3': 'Privatization',
                   'soru4': 'Death Penalty',
                   'soru5': 'Neutral Journalism',
                   'soru6': 'Alcohol Prohibition',
                   'soru7': 'Secularism',
                   'soru8': 'Abortion Ban',
                   'soru9': 'Ohal',
                   'soru10': 'New Party'
}, inplace=True)

df_other.rename(columns={'soru1': 'Economic Status',
                   'soru2': 'Education Reform',
                   'soru3': 'Privatization',
                   'soru4': 'Death Penalty',
                   'soru5': 'Neutral Journalism',
                   'soru6': 'Alcohol Prohibition',
                   'soru7': 'Secularism',
                   'soru8': 'Abortion Ban',
                   'soru9': 'Ohal',
                   'soru10': 'New Party'
}, inplace=True)

df.rename(columns={'soru1': 'Economic Status',
                   'soru2': 'Education Reform',
                   'soru3': 'Privatization',
                   'soru4': 'Death Penalty',
                   'soru5': 'Neutral Journalism',
                   'soru6': 'Alcohol Prohibition',
                   'soru7': 'Secularism',
                   'soru8': 'Abortion Ban',
                   'soru9': 'Ohal',
                   'soru10': 'New Party'
}, inplace=True)
```

```python
# Save separate dataframes
df_both.to_csv(r'C:\Users\Ebo\Projetos Personais\Github Repositories\Political_Opinions\data\02_leftright.csv', index=False)
df_other.to_csv(r'C:\Users\Ebo\Projetos Personais\Github Repositories\Political_Opinions\data\03_other.csv', index=False)
df.to_csv(r'C:\Users\Ebo\Projetos Personais\Github Repositories\Political_Opinions\data\01_wrangled.csv', index=False)
```
